```python
import pandas as pd
from sqlalchemy import create_engine

demographic_df = pd.read_csv("../Resources/demographic.csv")
flavor_df = pd.read_csv("../Resources/flavor.csv")
```

```python
coffee_df = pd.concat([demographic_df, flavor_df], axis=1, join='inner')
coffee_df
```

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'bean_juice_db'

rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
demographic_df.to_sql(name='demographic', con=engine, if_exists='append', index=False)
flavor_df.to_sql(name='flavor', con=engine, if_exists='append', index=False)
coffee_df.to_sql(name='coffee', con=engine, if_exists='append', index=False)
```
